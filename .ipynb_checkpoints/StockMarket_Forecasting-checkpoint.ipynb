#### Stock Market Prediction and Forecasting Using LSTM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
df = pd.read_csv('AAPL.csv',index_col=0)
df.head()
```

```python
df1 = df.reset_index()[['date','close']]
df1.shape
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(x=df1['date'],y=df1['close'])
plt.title('Stock Price of Apple Over the Years')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(6))
plt.show()
```

```python
df1 = df.reset_index()['close']
```

##### LSTM are sensitive to the scale of the data. so we apply MinMax scaler 

```python
scaler = MinMaxScaler(feature_range=(0,1))
df1 = scaler.fit_transform(np.array(df1).reshape(-1,1))
```

```python
df1.shape
```

```python
df1
```

```python
### Split the Data
## In time series data we split the data based on date
training_size = int(len(df1)*0.65)
test_size = len(df1) - training_size
train_data, test_data = df1[0:training_size,:],df1[training_size:len(df1),:1]
```

```python
len(train_data)
```

```python
len(test_data)
```

```python
def splitData(dataset,time_step=1):
    Xdata,Ydata = [], []
    for i in range(len(dataset) - time_step - 1):
        values = dataset[i:(i+time_step),0]
        Xdata.append(values)
        Ydata.append(dataset[i+time_step,0])
    return np.array(Xdata), np.array(Ydata)
```

```python
time_step = 100
X_train,y_train = splitData(train_data,time_step)
X_test,y_test = splitData(test_data,time_step)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

##### Reshape input to be [samples, time steps, features] which is required for LSTM

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

#### Model Building:

```python
model = Sequential()
model.add(LSTM(50,return_sequences=True, input_shape=(100,1)))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=64,verbose=1)
```

#### Prediction:

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
##Transformback to original form (As we scale the data using MinMax so have to unscale it)
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
```

```python
y_train1 = y_train.reshape(-1,1)
y_test1 = y_test.reshape(-1,1)
```

```python
y_train1 = scaler.inverse_transform(y_train1)
y_test1 = scaler.inverse_transform(y_test1)
```

##### Performance metrics:

```python
# RMSE
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train1,train_predict))
```

```python
### Test Data RMSE
math.sqrt(mean_squared_error(y_test1,test_predict))
```

```python
import numpy as np
import matplotlib.pyplot as plt

look_back = 100  # Number of previous time steps to use for prediction

# Create a new array filled with NaNs, with the same shape as df1
trainPredictPlot = np.empty_like(df1)
trainPredictPlot[:, :] = np.nan

# Fill in the 'trainPredictPlot' array with the predictions made during training
# The predictions start from 'look_back' index and go up to 'len(train_predict) + look_back'
trainPredictPlot[look_back:len(train_predict) + look_back, :] = train_predict

# Create a new array filled with NaNs, with the same shape as df1
testPredictPlot = np.empty_like(df1)
testPredictPlot[:, :] = np.nan

# Fill in the 'testPredictPlot' array with the predictions made during testing
# The predictions start from 'len(train_predict) + (look_back * 2) + 1' index
# and go up to 'len(df1) - 1'
testPredictPlot[len(train_predict) + (look_back * 2) + 1:len(df1) - 1, :] = test_predict

plt.figure(figsize=(14,7))
# Plot the original data and the predicted values
plt.plot(scaler.inverse_transform(df1))  # Plot the original data (scaled back to the original values)
plt.plot(trainPredictPlot)  # Plot the training set predictions
plt.plot(testPredictPlot)   # Plot the test set predictions
plt.show()
```

##### Predict Next 30 days:
To do that we need last 100 days data as we set step_size to 100 means that the model predict the value based on last 100 records. As test size have 441 size so we start from 341.

```python
x_input = test_data[341:].reshape(1,-1)
x_input.shape
```

```python
temp_input = list(x_input)
temp_input = temp_input[0].tolist()
temp_input
```

##### Predict 30 days:

```python
output = []  # List to store the predicted values
n_steps = 100  # Number of time steps to consider as input
i = 0 

while i < 30:  # Perform the following steps 30 times (or until i reaches 30)
    if len(temp_input) > 100:
        # If the length of temp_input is greater than 100, create an input sequence for prediction
        x_input = np.array(temp_input[1:])  # Remove the first element from temp_input as it is > 100 and we only need last 100 records
        x_input = x_input.reshape(1, -1)  # Reshape the input sequence to have one sample and multiple features
        x_input = x_input.reshape((1, n_steps, 1))  # Reshape the input to match the model's expected input shape
        yhat = model.predict(x_input, verbose=0)  # Make a prediction using the model
        temp_input.extend(yhat[0].tolist())  # Extend temp_input with the predicted value
        temp_input = temp_input[1:]  # Remove the first element from temp_input
        output.extend(yhat.tolist())  # Append the predicted value to the output list
        i += 1  # Increment the counter
    else:
        # If the length of temp_input is not greater than 100, create an initial input sequence for prediction
        x_input = x_input.reshape((1, n_steps, 1))  # Reshape the input to match the model's expected input shape
        yhat = model.predict(x_input, verbose=0)  # Make a prediction using the model
        temp_input.extend(yhat[0].tolist())  # Extend temp_input with the predicted value
        output.extend(yhat.tolist())  # Append the predicted value to the output list
        i += 1  # Increment the counter

print(output)  # Print the list of predicted values
```

```python
day_new=np.arange(1,101)
day_pred=np.arange(101,131)
```
